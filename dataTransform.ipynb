```python
import pandas as pd
```

```python
import numpy as np
```

```python
foglio3 = pd.read_csv("./foglio3.csv")
```

```python
foglio3.head(20)
```

```python
foglio2 = pd.read_csv("./copyofFoglio2.csv")
```

```python
foglio2.head()
```

```python
final_dataset = pd.DataFrame()
for i in range(1, 110):
    tmp = foglio2.copy()
    tmp['resp.id'] = i
    tmp['choice'] = 0
    for qid in range(1, 14):
        choice_id = foglio3[foglio3['ID']==i][f"Q{qid}"].item()
        idx = ((tmp['Question']==qid) & (tmp['Unnamed: 0']==choice_id))
        tmp.loc[idx, 'choice'] = 1
    final_dataset = final_dataset.append(tmp)
final_dataset.head(40)
```

```python
final_dataset.to_csv("./final.csv", index_label = False)
```
